```python
import pandas as pd
```

```python
# Cargar el archivo Excel
ruta_excel = "C:/Users/ferna/OneDrive - Universitat de les Illes Balears/PER/Avicultura/categorias analisis avicultura.xlsx"

# Obtener los nombres de todas las hojas del archivo Excel
analisis_avicultura = pd.ExcelFile(ruta_excel).sheet_names
```

```python
# Crear un diccionario para almacenar los dataframes generados
dfs = {}

# Bucle para leer cada hoja y crear los dataframes
for explotacion in analisis_avicultura:
    # Cargar los datos de la hoja actual
    datos_aves = pd.read_excel(ruta_excel, sheet_name=explotacion)

    # Crear los diferentes dataframes con las columnas y filas especificadas
    dfs[f"{explotacion}_ECREA"] = datos_aves.iloc[0:43, 0:3]
    dfs[f"{explotacion}_RECAN"] = datos_aves.iloc[:, 5:7]
    dfs[f"{explotacion}_SAU"] = datos_aves.iloc[0:3, 10:13]
    dfs[f"{explotacion}_Mano_Obra_asalariada"] = datos_aves.iloc[0:5, 16:25]
    dfs[f"{explotacion}_Mano_Obra_no_asalariada"] = datos_aves.iloc[0:5, 25:33]

# Imprimir las claves del diccionario generado
print("Claves generadas en el diccionario dfs:", dfs.keys())
```

```python
# Realizar el merge de los dataframes generados, como en R con left_join
datos_ecrea = dfs['Explotación_1 _ECREA'].merge(dfs['Explotación_2_ECREA'], 
                                               on=["Categoria", "Contabilidad"], 
                                               suffixes=('_Explotacion1', '_Explotacion2'))

datos_ecrea = datos_ecrea.merge(dfs['Explotación_3_ECREA'], 
                                on=["Categoria", "Contabilidad"], 
                                suffixes=('', '_Explotacion3'))

datos_ecrea = datos_ecrea.merge(dfs['Explotación_4_ECREA'], 
                                on=["Categoria", "Contabilidad"], 
                                suffixes=('', '_Explotacion4'))
datos_ecrea = datos_ecrea.drop_duplicates()
datos_ecrea
```

```python
# Repetir el proceso para RECAN y Préstamos
datos_recan = dfs['Explotación_1 _RECAN'].merge(dfs['Explotación_2_RECAN'], 
                                               on="Variable", 
                                               suffixes=('_Explotacion1', '_Explotacion2'))

datos_recan = datos_recan.merge(dfs['Explotación_3_RECAN'], 
                                on="Variable", 
                                suffixes=('', '_Explotacion3'))

datos_recan = datos_recan.merge(dfs['Explotación_4_RECAN'], 
                                on="Variable", 
                                suffixes=('', '_Explotacion4'))

datos_recan = datos_recan.drop_duplicates()
datos_recan
```
